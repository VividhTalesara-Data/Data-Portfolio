
# Pandas Data Visualization Exercise


```python
import pandas as pd
import matplotlib.pyplot as plt
df3 = pd.read_csv('df3')
%matplotlib inline
```

```python
df3.info()
```

```python
df3.head()
```

```python
df3.plot.scatter(x='a',y='b',c='red',s=50,figsize=(12,3))
```

```python
df3['a'].plot.hist(color='blue')
```

```python
plt.style.use('ggplot')
df3['a'].plot.hist(bins=25, alpha=0.5)
```

```python
df3[['a','b']].plot.box()
```

```python
df3['d'].plot(kind='kde')
```

```python
df3['d'].plot(kind='density', lw=4, ls='--')
```

```python
import matplotlib.pyplot as plt
f = plt.figure()
df3.iloc[0:30].plot.area(alpha=0.4, ax=f.gca())
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.show()
```
