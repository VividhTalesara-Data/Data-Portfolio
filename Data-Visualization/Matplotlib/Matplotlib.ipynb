```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.2,0.5,.2,.2])
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
plt.plot(x,y, 'red')
plt.xlabel('x')
plt.ylabel('y')
axes1.set_xlim([0,100])
axes1.set_ylim([0,200])

axes2 = fig.add_axes([0.2,0.5,.2,.2])
plt.plot(x,y, 'red')
plt.xlabel('x')
plt.ylabel('y')
axes2.set_xlim([0,100])
axes2.set_ylim([0,200])
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
plt.plot(x,z, 'b')
plt.xlabel('x')
plt.ylabel('z')
axes1.set_xlim([0,100])
axes1.set_ylim([0,10000])

axes2 = fig.add_axes([0.2,0.5,.4,.4])
plt.plot(x,y, 'b')
plt.xlabel('x')
plt.ylabel('y')
plt.title('zoom')
axes2.set_xlim([20,22])
axes2.set_ylim([30,50])
```

```python
plt.subplots(nrows=1, ncols=2)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x, y, color="blue", lw=3, ls='--')
axes[1].plot(x,z, color="red", lw=3, ls='-')

plt.tight_layout()
```
